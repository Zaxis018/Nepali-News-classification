```python
# Python program to print all heading tags
import requests
from bs4 import BeautifulSoup
import pandas as pd

a={}
key = 'title'
for i in range(1,49):
    t = str(i)
    url = 'https://gorkhapatraonline.com/categories/education?page='+ t
    

    request = requests.get(url)

    Soup = BeautifulSoup(request.text, 'lxml')

# creating a list of all common heading tags needed
    heading_tags = ["h2"]
    a.setdefault(key, [])
    for tags in Soup.find_all(heading_tags):
        a[key].append(tags.text.strip())

df = pd.DataFrame(a)
print(df)
df.to_csv('title.csv', index = False)
```
