```python
import numpy as np
X_input_ids = np.zeros((len(df), 512))
X_attn_masks = np.zeros((len(df), 512))
```

```python
from tqdm.auto import tqdm

def generate_training_data(df, ids, masks, tokenizer):
    for i, text in tqdm(enumerate(df['Description'])):
        tokenized_text = tokenizer.encode_plus(
            text,
            max_length=512,
            truncation=True,
            padding='max_length',
            add_special_tokens=True,
            return_tensors='tf'
        )
        ids[i, :] = tokenized_text.input_ids
        masks[i, :] = tokenized_text.attention_mask
    return ids, masks
```

```python
X_input_ids, X_attn_masks = generate_training_data(df, X_input_ids, X_attn_masks, tokenizer)
```

```python
X_input_ids
```

```python
X_attn_masks
```

```python
labels = np.zeros((len(df), 2))
labels.shape
```

```python
labels[np.arange(len(df)), df['label'].values] = 1
```

```python
# creating a data pipeline using tensorflow dataset utility, creates batches of data for easy loading...
dataset = tf.data.Dataset.from_tensor_slices((X_input_ids, X_attn_masks, labels))
dataset.take(1) # one sample data
```

```python
def DatasetMapFunction(input_ids, attn_masks, labels):
    return {
        'input_ids': input_ids,
        'attention_mask': attn_masks
    }, labels
```

```python
dataset = dataset.map(DatasetMapFunction) # converting to required format for tensorflow dataset
```

```python
dataset.take(1)
```

```python
dataset = dataset.shuffle(10000).batch(32, drop_remainder=True)
```

```python
p = 0.8
train_size = int((len(df)/32)*p) # for each 16 batch of data we will have len(df)//16 samples, take 80% of that for train.
```

```python
train_dataset = dataset.take(train_size)
test_dataset = dataset.skip(train_size)
```

```python
len(test_dataset)
```

```python
input_ids = tf.keras.layers.Input(shape=(512,), name='input_ids', dtype='int32')
attn_masks = tf.keras.layers.Input(shape=(512,), name='attention_mask', dtype='int32')

bert_model.bert.trainable = False
bert_embds = bert_model.bert(input_ids, attention_mask=attn_masks)[1] # 0 -> activation layer (3D), 1 -> pooled output layer (2D)


#hidden layers
intermediate_layer = tf.keras.layers.Dense(512, activation='relu', name='intermediate_layer')(bert_embds)
dropout_layer = tf.keras.layers.Dropout(0.1)(intermediate_layer)

intermediate_layer2 = tf.keras.layers.Dense(256, activation='relu', name='intermediate_layer2')(dropout_layer)
dropout_layer2 = tf.keras.layers.Dropout(0.1)(intermediate_layer2)

output_layer = tf.keras.layers.Dense(2, activation='sigmoid', name='output_layer')(dropout_layer2)
model = tf.keras.Model(inputs=[input_ids, attn_masks], outputs=output_layer)
model.summary()
```

```python
from keras.optimizers import Adam
optimizer = Adam(learning_rate=1e-3)

METRICS = [
      tf.keras.metrics.BinaryAccuracy(name='accuracy'),
      tf.keras.metrics.Precision(name='precision'),
      tf.keras.metrics.Recall(name='recall')
]

model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=METRICS)
```

```python
history = model.fit(train_dataset, validation_data=test_dataset,epochs=1)
```

```python
model.save('model.h5')
```

```python
#checking on new data
model = tf.keras.models.load_model('model.h5')

tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')

def prepare_data(input_text, tokenizer):
    token = tokenizer.encode_plus(
        input_text,
        max_length=512,
        truncation=True,
        padding='max_length',
        add_special_tokens=True,
        return_tensors='tf'
    )
    return {
        'input_ids': tf.cast(token.input_ids, tf.float64),
        'attention_mask': tf.cast(token.attention_mask, tf.float64)
    }

def make_prediction(model, processed_data, classes=['True', 'Fake']):
    probs = model.predict(processed_data)[0]
    print(f"Prediction confidence: {probs}")
    return classes[np.argmax(probs)]
```

```python

input_text = input('Enter Nepali news here: ')
processed_data = prepare_data(input_text, tokenizer)
result = make_prediction(model, processed_data=processed_data)
probs = model.predict(processed_data)[0]
if result=='Fake':
  print(probs[1])
else:
  print(probs[0])

print(f"Predicted result: {result}")
```
